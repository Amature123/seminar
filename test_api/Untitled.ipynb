```python
from vnstock import Vnstock
stock = Vnstock().stock(symbol='TCB', source='vci')
in_day = stock.quote.intraday(page_size=5, show_log=False)
print(in_day)
records = []
for _, row in in_day.iterrows():
    record = {
        "id": row["id"],
        "symbol": 'TCB',
        "price": row["price"],    # giá khớp lệnh
        "volume": row["volume"],    # khối lượng khớp lệnh
        "type": row["match_type"],            # buy / sell
        "time": row["time"].strftime("%Y-%m-%d %H:%M:%S"),    
    }
    records.append(record)
```

```python
import re
import pandas as pd
def extract_company_info(symbol)->pd.DataFrame:
    company = Company(symbol=symbol,source='VCI')
    return company.overview()

def extract_history(symbol):
    text = extract_company_info(symbol)['history'].iloc[0]
    pattern = r"-\s*(Ngày\s\d{2}/\d{2}/\d{4}|Năm\s\d{4}):\s*(.*?);"
    matches = re.findall(pattern, text)
    df = pd.DataFrame(matches, columns=["date", "detail"])
    return df
df=extract_history('ACB')

df.head()
```
